## Data Cleaning

When reviewing the column names, you may notice that we have different cases (upper and lower case). Some had spaces in the beginning of the name or maybe they have multiple spaces in between. Let's clean this first so that we could better reference them as we go along with the challenge.

```python
life_df.columns = life_df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')','')
```

Let us review the change on the column names and check the number of unique values.

```python
life_df.nunique()
```

**YES!** This is easier and pleasing to the eyes. No spaces and all in lower case. This will cause less confusion what column we're playing with.

Now it's PLAYTIME! 
Next stop in the journey is filling in the missing data. I want to look at **Life Expectancy** first. Since we have low percentage of missing data, we can take the *mean* on the account.

```python
life_df["life_expectancy"].fillna(life_df["life_expectancy"].mean(), inplace=True)
```

Cool Beans! Now what's next?
How about the following data, adult_mortality, alcohol, bmi, total_expenditure, and income_composition_of_resources. Since the data missing is relative small, let's take the **mean** to fill in the values for those datasets.

```python
life_df["adult_mortality"].fillna(life_df["adult_mortality"].mean(), inplace=True)
life_df["alcohol"].fillna(life_df["alcohol"].mean(), inplace=True)
life_df["bmi"].fillna(life_df["bmi"].mean(), inplace=True)
life_df["total_expenditure"].fillna(life_df["total_expenditure"].mean(), inplace=True)
life_df["income_composition_of_resources"].fillna(life_df["income_composition_of_resources"].mean(), inplace=True)
```

I know what you are thinking... how about the data with large percentage of missing data?
We would **drop** them. Why? Using the mean for the missing values would change the whole data itself. It is best to remove them from the data to keep the data's integrity.

```python
life_df.dropna(how='any', subset=['population','hepatitis_b','gdp'], inplace=True)
```

How are we doing with missing data? Can we check where we at? Yes, we can!

```python
life_df.isnull().sum()*100/life_df.isnull().count()
```

Looking at the columns one more time, there are some columns that we could remove all together aka *nonsense values*. These are columns that would not change our overall analysis in finding what affects life expectancy. Data like **year, thinness__1-19_years and thinness_5-9_years** are extra information that we can discard or ignore for it will not affect our overall analysis.

### Data Exploration

After all that cleaning we need to check for data points that would skew our data from one to another, aka, outliers. In this part of our journey we will use the `boxplot` to view if we have any outliers that would possible change our data. Not all data was plotted for I want to only look at values that would possible have a correlation with *life expectancy*.

```python
plt.figure(figsize = (30,30))

plt.subplot(4,3,1) 
plt.boxplot(life_df["life_expectancy"]) 
plt.title("Life Expectancy")

plt.subplot(4,3,2)
plt.boxplot(life_df["percentage_expenditure"])
plt.title("Percentage Expenditure")

plt.subplot(4,3,3)
plt.boxplot(life_df["gdp"])
plt.title("GPD")


plt.subplot(4,3,4)
plt.boxplot(life_df["schooling"])
plt.title("schooling")

plt.subplot(4,3,5)
plt.boxplot(life_df["hepatitis_b"])
plt.title("Hepatitis B")


plt.subplot(4,3,6)
plt.boxplot(life_df["measles"])
plt.title("Measles")


plt.subplot(4,3,7)
plt.boxplot(life_df["polio"])
plt.title("Polio")

plt.subplot(4,3,8)
plt.boxplot(life_df["diphtheria"])
plt.title("Diphtheria")

plt.subplot(4,3,9)
plt.boxplot(life_df["hiv/aids"])
plt.title("HIV / Aids")


plt.subplot(4,3,10)
plt.boxplot(life_df["adult_mortality"])
plt.title("Adult Mortality")

plt.subplot(4,3,11)
plt.boxplot(life_df["alcohol"])
plt.title("alcohol")

plt.subplot(4,3,12)
plt.boxplot(life_df["bmi"])
plt.title("BMI")

plt.show()
```

As expected, **WE DO have OUTLIERS!** What do we do now? Let's winsorize them to limit extreme values in the statistical data and to reduce the effect of outliers. Let's do that to all exception for BMI. That data looks ok.

```python
from scipy.stats.mstats import winsorize

life_df["wlife_expectancy"] = winsorized_views = winsorize(life_df["life_expectancy"], (0, 0.10))
life_df["wpercentage_expenditure"] = winsorized_views = winsorize(life_df["percentage_expenditure"], (0, 0.10))
life_df["wgdp"] = winsorized_views = winsorize(life_df["gdp"], (0, 0.10))

life_df["wschooling"] = winsorized_views = winsorize(life_df["schooling"], (0, 0.10))

life_df["wmeasles"] = winsorized_views = winsorize(life_df["measles"], (0, 0.10))
# removed hepatitis, polio, diphtheria  because winsorized has a 1:1 correlation

life_df["whiv/aids"] = winsorized_views = winsorize(life_df["hiv/aids"], (0, 0.10))

life_df["wadult_mortality"] = winsorized_views = winsorize(life_df["adult_mortality"], (0, 0.10))
life_df["walcohol"] = winsorized_views = winsorize(life_df["alcohol"], (0, 0.10))
```

*side note: The section above was revisted after plotting the correlation heat map.*

Let's review our data with another set of `boxplot`.

```python
plt.figure(figsize = (30,30))

plt.subplot(4,3,1) 
plt.boxplot(life_df["wlife_expectancy"], whis = 5) 
plt.title("Life Expectancy")

plt.subplot(4,3,2)
plt.boxplot(life_df["wpercentage_expenditure"], whis = 5)
plt.title("Percentage Expenditure")

plt.subplot(4,3,3)
plt.boxplot(life_df["wgdp"], whis = 5)
plt.title("GPD")

plt.subplot(4,3,4)
plt.boxplot(life_df["wschooling"], whis = 5)
plt.title("schooling")

plt.subplot(4,3,5)
plt.boxplot(life_df["wmeasles"], whis = 15)
plt.title("Measles")

plt.subplot(4,3,6)
plt.boxplot(life_df["whiv/aids"], whis = 15)
plt.title("HIV / Aids")

plt.subplot(4,3,7)
plt.boxplot(life_df["wadult_mortality"], whis = 5)
plt.title("Adult Mortality")

plt.subplot(4,3,8)
plt.boxplot(life_df["walcohol"], whis = 5)
plt.title("alcohol")

plt.show()
```

### Developed vs Developing

It make sense that the larger the population, the more income you have. I'm surprise to see that the total expenditure between developed and developing countries are not that far apart. *I expect that developed countries would spend more in making their countrymen healthy.*

How about death or mortality between adult and infants? How do they differ per status?

```python
life_df.groupby("status").count()[["wadult_mortality", "infant_deaths"]]
```

Now, let us review the **different deseases**. How do they stand against the developed and developing countries?

```python
life_df.groupby("status").mean()[["wmeasles", "hepatitis_b","polio","diphtheria","walcohol","whiv/aids"]]
```

In **developed countries**, they have a high rate of mortality due to Hepatitis B, Polio, Diphtherua and death cause by Alcohol. While in **developing countries**, they have high rate of death caused by Measles and HIV/Aids.

### View by Countries
Since we know what would likely cause mortality per status, let's name the different countries. Using `pivot_table`, let's view what countries are named *developed* and *developing*

```python
developed_df1=life_df[life_df.status == 'Developed'].pivot_table( index=['country'], 
                                                                 values=["wlife_expectancy"], aggfunc="mean")
developed_df2 = developed_df1.reset_index()
developed_df2.sort_values(by=['wlife_expectancy'], inplace=True)
developed_df2.plot.bar(x='country', y='wlife_expectancy')
plt.show()
```

Our top countries that we could also named, most healthy places are: *Austria, Malta, Spain, Ireland, Australia, Italy, Netherlands and Sweden*. I heard there's a lot of happy people in the Netherlands and Sweden. Let's add that to a list of places to visit.

Curiosity wants to know what countries are named under the developing status. Let's look at them using the same method.

```python
developing_df1 = life_df[life_df.status == 'Developing'].pivot_table(index=['country'], 
                                                                  values=["wlife_expectancy"], aggfunc="mean")
developing_df2 = developing_df1.reset_index()

developing_df2.sort_values(by=['wlife_expectancy'], inplace=True)

plt.rcParams['figure.figsize'] = [20, 5]

developing_df2.plot.bar(x='country', y='wlife_expectancy')
plt.show()
```

A lot more data for sure but countries with high mortality rate is in *Sierra Leone, Lesotho, Zimbabwe, Malawi and Angola*
